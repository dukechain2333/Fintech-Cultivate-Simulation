```python
import numpy as np

```

```python
data = np.load('data.npy')
data
```

```python
financeStudent = 0
CSStudent = 0
fintechStudent = 0

for d in data:
    if d[0][0]==0:
        CSStudent+=1
    elif d[0][1]==0:
        financeStudent+=1
    else:
        fintechStudent+=1
```

```python
finalFinance = []
finalCS = []
finalFintech = []
```

```python
tmpData = []
for d in data:
    tmpData.append(d[-1])
tmpData
```

```python
for t in tmpData:
    if t[0]==0:
        finalCS.append(t)
    elif t[1]==0:
        finalFinance.append(t)
    else:
        finalFintech.append(t)
```

```python
finalCS = np.array(finalCS)
finalCS
```

```python
finalFinance = np.array(finalFinance)
finalFinance
```

```python
finalFintech = np.array(finalFintech)
finalFintech
```

```python
np.save('FinTechFinal', finalFintech, allow_pickle=True, fix_imports=True)
np.save('FinanceFinal', finalFinance, allow_pickle=True, fix_imports=True)
np.save('CSFinal', finalCS, allow_pickle=True, fix_imports=True)
```
